### Importing necessary modules

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import normalize
import scipy.cluster.hierarchy as shc
%matplotlib inline
```

```python
# Reading the dataset
customers = pd.read_csv('data.csv')
customers.head()
```

```python
#Scaling the dataset by normalizing
customers_scaled = normalize(customers)
customers_scaled = pd.DataFrame(customers_scaled, columns=customers.columns)
customers_scaled.head()
```

```python
# Plotting the data
plt.figure(figsize=(16, 8))  
plt.title("Dendrogram")  
dendogram = shc.dendrogram(shc.linkage(customers_scaled, method='ward'))
```
