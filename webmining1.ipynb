```python
# define punctuation
punctuations = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''
text = "Nick likes to play football, however he is not too fond of tennis. Hello!!!, he said ---and went."

# remove punctuation from the string
no_punct = ""
for char in text:
    if char not in punctuations:
        no_punct = no_punct + char

# display the unpunctuated string
print(no_punct)
text_tokens = word_tokenize(no_punct)
tokens_without_sw = [word for word in text_tokens if not word in stopwords.words()]
print(tokens_without_sw)
```

```python
from math import sqrt
def Unary(x):
    return (x-1)*'0'+'1'

def prime(x):
    if(x == 1):
        return False
    for i in range(2, int(sqrt(x)+1)):
        if(x % i == 0):
            return False
    return True

def unarydecode(x):
    x = Unary(x)
    return x.count('0')+1

for i in range(1, 100):
    if(prime(i)):
        print('Encoding')
        print(i, ':', Unary(i))
        print('Decoding')
        print(Unary(i), unarydecode(i))
```
