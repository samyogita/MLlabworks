```python
#reading the dataset
def read_data(filename):

    with open(filename,'r') as csvfile:
        datareader = csv.reader(csvfile)
        metadata = next(datareader)
        traindata=[]
        for row in datareader:
            traindata.append(row)

    return (metadata, traindata)
```

```python
#splitting the data into training and testing
def splitDataset(dataset, splitRatio):
    trainSize = int(len(dataset) * splitRatio)
    trainSet = []
    testset = list(dataset)
    i=0
    while len(trainSet) < trainSize:
        trainSet.append(testset.pop(i))
    return [trainSet, testset]
```

```python
#classifying the data 
def classify(data,test):

    total_size = data.shape[0]
    print("\n")
    print("Size of training data=",total_size) # total size of the training set
    print("Size of test data=",test.shape[0]) # test size of the training set

    total_yes = 0
    total_no = 0
    prob_yes = 0
    prob_no = 0
    print("\n")
    print("[0-not spam \t 1-spam]")
    print("Target Total count [0/1] Probability")

    for x in range(data.shape[0]):
        if data[x,data.shape[-1]-1] == '1':
            total_yes +=1
        if data[x,data.shape[-1]-1] == '0':
            total_no +=1

    prob_yes=total_yes/total_size
    prob_no= total_no / total_size

    print('1',"\t",total_yes,"\t",prob_yes)
    print('0',"\t",total_no,"\t",prob_no)


    prob0 =np.zeros((test.shape[1]-1))
    prob1 =np.zeros((test.shape[1]-1))
    accuracy=0
    print("\n")
    print("Email Prediction(NB) Actual Target")

    for t in range(test.shape[0]):
        for k in range (test.shape[1]-1):
            count1=count0=0
            for j in range (data.shape[0]):
                #how many times appeared with no
                if test[t,k] == data[j,k] and data[j,data.shape[1]-1]== '0':
                    count0+=1
                #how many times appeared with yes
                if test[t,k]==data[j,k] and data[j,data.shape[1]-1]=='1':
                    count1+=1
            prob0[k]=count0/total_no
            prob1[k]=count1/total_yes

        prob_no=prob_no
        prob_yes=prob_yes
        for i in range(test.shape[1]-1):
            prob_no=prob_no*prob0[i]
            prob_yes=prob_yes*prob1[i]
        if prob_no>prob_yes:
            predict=0
        else:
            predict=1

        print(t+1,"\t",predict,"\t    ",test[t,test.shape[1]-1])
        if predict == test[t,test.shape[1]-1]:
            accuracy+=1
    final_accuracy=(accuracy/test.shape[0])*100
    print("accuracy",final_accuracy,"%")
    return

metadata,traindata= read_data("emails.csv")
splitRatio=0.99
trainingset, testset=splitDataset(traindata, splitRatio)
training=np.array(trainingset)
testing=np.array(testset)
classify(training,testing)
```
